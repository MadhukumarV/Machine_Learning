```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('F:/DATA ANALYSIS/Machine Learnig/Complete Machine Learning with Python/Data Files/3. ST Academy - Decision Trees resource files/Movie_classification.csv',header = 0)
```

```python
df.head()
```

```python
df.info()
```

### Missing Value Imputation

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value = df['Time_taken'].mean(),inplace = True)
```

```python
df.info()
```

## Dummy variable

```python
df.head()
```

```python
df =pd.get_dummies(df,columns =['3D_available','Genre'],drop_first = True)
```

```python
df.head()
```

## x-y Split

```python
X = df.loc[:,df.columns!="Start_Tech_Oscar"]
type(X)
```

```python
X.head()
```

```python
X.shape
```

```python
y = df["Start_Tech_Oscar"]
type(y)
```

```python
y.shape
```

### Test Train split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.2 ,random_state = 0)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

### Training Classification tree

```python
from sklearn import tree
clftree = tree.DecisionTreeClassifier(max_depth = 3)
```

```python
clftree.fit(X_train,y_train)
```

### predict value using train models

```python
y_train_pred =  clftree.predict(X_train)
y_test_pred  = clftree.predict(X_test)
```

```python
y_test_pred  
```

### Model Performance

```python
from sklearn.metrics import  accuracy_score,confusion_matrix
```

```python
confusion_matrix(y_train,y_train_pred)
```

```python
confusion_matrix(y_test,y_test_pred)
```

```python
accuracy_score(y_test,y_test_pred)
```

### Plotting Decision tree

```python
dot_data = tree.export_graphviz(clftree,out_file = None,feature_names =X_train.columns,filled =True)
```

### Controlling Tree Growth

```python
clftree2 = tree.DecisionTreeClassifier(min_samples_leaf=20,max_depth = 4)
clftree2.fit(X_train,y_train)
dot_data = tree.export_graphviz(clftree2,out_file =None,feature_names = X_train.columns,filled =True)
graph2 = pydotplus.graph_from_dot_data(dot_data)
Image(graph2.create_png())
```

```python
accuracy_score(y_test,clftree2.predict(X_test))
```

### Bagging in python

```python
from sklearn import tree
```

```python
clftree6 = tree.DecisionTreeClassifier()
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag_clf  = BaggingClassifier(estimator=clftree6,n_estimators = 1000,bootstrap = True,n_jobs =-1,random_state =42)
```

```python
bag_clf.fit(X_train,y_train)
```

```python
confusion_matrix(y_test,bag_clf.predict(X_test))
```

```python
accuracy_score(y_test,bag_clf.predict(X_test))
```

### Random forest 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_clf = RandomForestClassifier(n_estimators= 1000,n_jobs =-1,random_state = 42)
```

```python
rf_clf.fit(X_train,y_train)
```

```python
confusion_matrix(y_test,rf_clf.predict(X_test))
```

```python
accuracy_score(y_test,rf_clf.predict(X_test))
```

### Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_clf1 = RandomForestClassifier(n_estimators= 250,random_state = 42)
```

```python
param_grid = {"max_features":[4,5,6,7,8,9,10],
              "min_samples_split":[2,3,10]}
```

```python
grid_search = GridSearchCV(rf_clf1,param_grid,n_jobs=-1,cv=5,scoring = 'accuracy')
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
cvrf_clf = grid_search.best_estimator_
```

```python
accuracy_score(y_test,cvrf_clf.predict(X_test))
```

```python
confusion_matrix(y_test,cvrf_clf.predict(X_test))
```
