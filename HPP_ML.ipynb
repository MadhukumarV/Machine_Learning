### Logistic Regression

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('F:/DATA ANALYSIS/Machine Learnig/Complete Machine Learning with Python/Data Files/2. ST Academy - Classification models resource files/Classification preprocessed data Python.csv',header=0)
df.head()
```

```python
X = df[['price']]
```

```python
y= df['Sold']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf_lrs =  LogisticRegression()
```

```python
clf_lrs.fit(X,y)
```

```python
clf_lrs.coef_
```

```python
clf_lrs.intercept_
```

```python
import statsmodels.api as sn
```

```python
X_cons = sn.add_constant(X)
```

```python
X_cons .head()
```

```python
import statsmodels.discrete.discrete_model as sm
```

```python
logit = sm.Logit(y,X_cons).fit()
```

```python
logit.summary()
```
